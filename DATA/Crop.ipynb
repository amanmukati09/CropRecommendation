```python
ds = pd.read_csv("Weather.csv")
ds.head()
```

```python
dc = pd.read_csv("Crop.csv")
dc
```

```python
print("We Will See Which Crops are suitable for our city :- ")

City = int(input("City : "))
Month = int(input("Month : "))

print("For city :"+ str(City) + " and month : " + str(Month))
print("Suitable Crops are :- ")

row = City*12 + Month - 1

h = ds["Humidity"][row]
t = ds["Temperature"][row]
p = ds["Precipatition"][row]*12
pp = ds["pptAnnual"][row]


for i in range(0,22):
    if((h >= dc["mH"][i] and h <= dc["MH"][i] ) and (t >= dc["mT"][i] and t <= dc["MT"][i]) and (dc["MR"][i] < pp )):        
        print("\n\nSoil Conditions ;- ")
        print("Soil Nitrogen : " + str(dc["mN"][i]) + " - " + str(dc["MN"][i]))
        print("Soil Phosphorus  : " + str(dc["mP"][i]) + " - " + str(dc["MP"][i]))
        print("Soil Pottassium : " + str(dc["mK"][i]) + " - " + str(dc["MK"][i]))
        print("Soil PH : " + str(dc["mPh"][i]) + " - " + str(dc["MPh"][i]))
        print("Based On Above Soil Conditions Crops Will Be :- ")
        print(dc["Crop"][i])
```
